```sql
CREATE SCHEMA IF NOT EXISTS ops;

CREATE TABLE IF NOT EXISTS ops.ingest_log (
  -- Identity
  target_table_fqn    STRING,
  layer               STRING,
  pipeline_name       STRING,
  event_level         STRING,

  -- Run-level
  run_id              STRING,
  pipeline_run_ts     TIMESTAMP,

  -- Auto Loader state (debuggable)
  checkpoint_location STRING,
  schema_location     STRING,
  batch_id            BIGINT,

  -- File identity
  source_file         STRING,

  -- Metrics
  row_count           BIGINT,
  rows_rescued        BIGINT,

  -- Timing
  started_ts          TIMESTAMP,
  finished_ts         TIMESTAMP,

  -- Status & diagnostics
  status              STRING,   -- SUCCESS / FAILED
  message             STRING,
  error_class         STRING,

  -- Optional: future-proof hook for dataset-specific fields
  context_json        STRING
)
USING DELTA;
```
